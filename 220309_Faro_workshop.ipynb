# Data pre-procesing and data cleaning

```python
################################# Data pre-procesing and data cleaning ######################################

#Drop rows that the post value is empty
#df.dropna(axis=0, how='any', thresh=None,subset=["sentence_wind"], inplace=True)


#change to lower case
df["post_en_preprocessed"] = df["post_en"].str.lower()

#emoji (IT REMOVES THE PERSIANS!!!!)
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'[^\x00-\x7F]+', '')

#user mention removes
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'@\S+', '')
                                
#url removes
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'(https|http)?:\/(\w|\.|\/|\?|\=|\&|\%)*\b','')
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'www\.\S+\.com','')

#Remove hashtags
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'#', ' #')
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'# ', '#')
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r',', ' ,')
# df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'#.*?(?=\s|$)','')

#punctuation
import string
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])# It will discard all punctuations
    return text_nopunct
df["post_en_preprocessed"] = df["post_en_preprocessed"].apply(lambda x: remove_punct(x))


#stop words removes
# Import stopwords with nltk.
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')
df["post_en_preprocessed"] = df["post_en_preprocessed"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))


#put an space after each full stop
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r'.', '. ')


#removes extra spaces
df["post_en_preprocessed"] = df["post_en_preprocessed"].str.replace(r' +', ' ')

print(df["post_en_preprocessed"].head())
# print(df)



# df.to_csv( "C:\TUDelft\Data_Collection_Windcatcher\literature\literature_full texts\220103literatureData_wind_sen_list.csv")
```

```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from pattern.en import singularize

### Define operations and others:
sentence_re = r'(?:(?:[A-Z])(?:.[A-Z])+.?)|(?:\w+(?:-\w+)*)|(?:\$?\d+(?:.\d+)?%?)|(?:...|)(?:[][.,;"\'?():-_`])'
lemmatizer = nltk.WordNetLemmatizer()
stemmer = nltk.stem.porter.PorterStemmer()
grammar = r"""
    NBAR:
        {<NN.*|JJ>*<NN.*>}  # Nouns and Adjectives, terminated with Nouns
        
    NP:
        {<NBAR>}
        {<NBAR><IN><NBAR>}  # Above, connected with in/of/etc...
"""
chunker = nltk.RegexpParser(grammar)


def leaves(tree):
    """Finds NP (nounphrase) leaf nodes of a chunk tree."""
    for subtree in tree.subtrees(filter = lambda t: t.label()=='NP'):
        yield subtree.leaves()

def normalise(word):
    """Normalises words to lowercase and stems and lemmatizes it."""
    word = word.lower()
    # word = stemmer.stem_word(word) #if we consider stemmer then results comes with stemmed word, but in this case word will not match with comment
    word = lemmatizer.lemmatize(word)
   
    return word


# stopwords = [‚Äòourselves‚Äô, ‚Äòhers‚Äô, ‚Äòbetween‚Äô, ‚Äòyourself‚Äô, ‚Äòbut‚Äô, ‚Äòagain‚Äô, ‚Äòthere‚Äô, ‚Äòabout‚Äô, ‚Äòonce‚Äô, ‚Äòduring‚Äô, ‚Äòout‚Äô, ‚Äòvery‚Äô, ‚Äòhaving‚Äô, ‚Äòwith‚Äô, ‚Äòthey‚Äô, ‚Äòown‚Äô, ‚Äòan‚Äô, ‚Äòbe‚Äô, ‚Äòsome‚Äô, ‚Äòfor‚Äô, ‚Äòdo‚Äô, ‚Äòits‚Äô, ‚Äòyours‚Äô, ‚Äòsuch‚Äô, ‚Äòinto‚Äô, ‚Äòof‚Äô, ‚Äòmost‚Äô, ‚Äòitself‚Äô, ‚Äòother‚Äô, ‚Äòoff‚Äô, ‚Äòis‚Äô, ‚Äòs‚Äô, ‚Äòam‚Äô, ‚Äòor‚Äô, ‚Äòwho‚Äô, ‚Äòas‚Äô, ‚Äòfrom‚Äô, ‚Äòhim‚Äô, ‚Äòeach‚Äô, ‚Äòthe‚Äô, ‚Äòthemselves‚Äô, ‚Äòuntil‚Äô, ‚Äòbelow‚Äô, ‚Äòare‚Äô, ‚Äòwe‚Äô, ‚Äòthese‚Äô, ‚Äòyour‚Äô, ‚Äòhis‚Äô, ‚Äòthrough‚Äô, ‚Äòdon‚Äô, ‚Äònor‚Äô, ‚Äòme‚Äô, ‚Äòwere‚Äô, ‚Äòher‚Äô, ‚Äòmore‚Äô, ‚Äòhimself‚Äô, ‚Äòthis‚Äô, ‚Äòdown‚Äô, ‚Äòshould‚Äô, ‚Äòour‚Äô, ‚Äòtheir‚Äô, ‚Äòwhile‚Äô, ‚Äòabove‚Äô, ‚Äòboth‚Äô, ‚Äòup‚Äô, ‚Äòto‚Äô, ‚Äòours‚Äô, ‚Äòhad‚Äô, ‚Äòshe‚Äô, ‚Äòall‚Äô, ‚Äòno‚Äô, ‚Äòwhen‚Äô, ‚Äòat‚Äô, ‚Äòany‚Äô, ‚Äòbefore‚Äô, ‚Äòthem‚Äô, ‚Äòsame‚Äô, ‚Äòand‚Äô, ‚Äòbeen‚Äô, ‚Äòhave‚Äô, ‚Äòin‚Äô, ‚Äòwill‚Äô, ‚Äòon‚Äô, ‚Äòdoes‚Äô, ‚Äòyourselves‚Äô, ‚Äòthen‚Äô, ‚Äòthat‚Äô, ‚Äòbecause‚Äô, ‚Äòwhat‚Äô, ‚Äòover‚Äô, ‚Äòwhy‚Äô, ‚Äòso‚Äô, ‚Äòcan‚Äô, ‚Äòdid‚Äô, ‚Äònot‚Äô, ‚Äònow‚Äô, ‚Äòunder‚Äô, ‚Äòhe‚Äô, ‚Äòyou‚Äô, ‚Äòherself‚Äô, ‚Äòhas‚Äô, ‚Äòjust‚Äô, ‚Äòwhere‚Äô, ‚Äòtoo‚Äô, ‚Äòonly‚Äô, ‚Äòmyself‚Äô, ‚Äòwhich‚Äô, ‚Äòthose‚Äô, ‚Äòi‚Äô, ‚Äòafter‚Äô, ‚Äòfew‚Äô, ‚Äòwhom‚Äô, ‚Äòt‚Äô, ‚Äòbeing‚Äô, ‚Äòif‚Äô, ‚Äòtheirs‚Äô, ‚Äòmy‚Äô, ‚Äòagainst‚Äô, ‚Äòa‚Äô, ‚Äòby‚Äô, ‚Äòdoing‚Äô, ‚Äòit‚Äô, ‚Äòhow‚Äô, ‚Äòfurther‚Äô, ‚Äòwas‚Äô, ‚Äòhere‚Äô, ‚Äòthan‚Äô]
#     stopwords = stopwords.words('english')
def acceptable_word(word):
    """Checks conditions for acceptable word: length, stopword. We can increase the length if we want to consider large phrase"""
    accepted = bool(3 <= len(word) <= 40
        and word.lower())# not in stopwords)
        
    return accepted


def get_terms(tree):
    for leaf in leaves(tree):
#         print(leaf)
#         for w,t in leaf:
#             print(t)
        term = [ normalise(w) for w,t in leaf if acceptable_word(w) and t == "NN" ]
        yield term

####################### Call the dataframework and apply the function ############################
# text = "yes good morning ladies and gentlemen heartily Welcome to the panels factory Special building from the thirties of the twentieth century The Van Nelle factory the cathedral of The new build on the edge of Rotterdam a monument of the progress and since 2014 UNESCO World Heritage Site From 1930 it was the factory of it "
df["attributes"] = ""
counter = -1
for text in df["post_en_preprocessed"]:
    counter = counter +1
#     words = sen.split(' ')
    print(counter, text, type(text))
   
    
    toks = nltk.regexp_tokenize(text, sentence_re)
######### singularize ###########
# print(toks, type(toks))
    toks2 = []

    for tok in toks:
#     print(tok)
#     for pluralForm in nounls:
      tok2= singularize(tok)
#       print("tok",tok)
      toks2.append(tok2)
#       print(toks2)
    postoks = nltk.tag.pos_tag(toks2)
#     print ("postoks", postoks)
    tree = chunker.parse(postoks)
    
# use the functions defined above :)
    terms = get_terms(tree)
    
# Add the results to a column in the dataset    
    list = []
    for term in terms:
      for word in term:
#         print (word)
        list.append(word)
#         print(list)
    df.at[counter,"attributes"] = list
########################### save to the a dataframe ######################        
# df.to_excel("C:\\TUDelft\\meeting_conferences_Journals\\workshop_Faro_social media\\processed_2022_3_9#faroportugal_hashtag_on_Insta_ScrapingData_ScrapeStorm.xlsx")   
```
